```python
import glob, os
import folium
from folium import Choropleth, Circle, Marker
from folium.plugins import HeatMap, MarkerCluster
from bs4 import BeautifulSoup

m_1 = folium.Map(location=[60.619702, 56.768939], tiles='openstreetmap', zoom_start=8)
mc = MarkerCluster()

os.chdir("data_xml")
for file in glob.glob("*.xml"):
    xml_doc = open(file, encoding='utf-8').read()
    soup = BeautifulSoup(xml_doc, 'lxml-xml')
    
    for item in soup.findAll('tab'):
        popup = item.find('COORD_W').text + ' ' + item.find('COORD_L').text
        tooltip = item.find('DTPV').text
        color = 'black'
        if (item.find('DTPV').text == 'Наезд на пешехода'):
            color = 'red'
        
        mc.add_child(Marker([item.find('COORD_W').text, item.find('COORD_L').text], icon=folium.Icon(color=color,icon_color='#FFFF00'), popup=popup, tooltip=tooltip))

m_1.add_child(mc)
# Display the map
m_1

```
